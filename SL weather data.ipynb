```python
df1 = df1.drop_duplicates()
```

```python
import pandas as pd

df1['time'] = pd.to_datetime(df1['time'])

# Filter records from 2019 to 2024 and for specific cities
cities = ['Colombo', 'Kandy', 'Matara','Deniyaya', 'Nuwara Eliya', 'Bibile', 'Monaragala', 'Kurunegala', 'Jaffna']
df2 = df1[(df1['time'].dt.year >= 2019) & (df1['time'].dt.year <= 2024) & (df1['city'].isin(cities))]

# Print the filtered DataFrame
df2
```
